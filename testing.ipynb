```python
cancer = load_breast_cancer()
df = pd.DataFrame(cancer.data, columns=cancer.feature_names).assign(
    target=cancer.target
)
df.columns = [c.replace(" ", "_") for c in df.columns]
df["target"] = df["target"].astype("category")
for col in df.columns:
    if col != "target":
        df[f"{col}_gt_75th_pctile"] = (
            (df[col] > df[col].quantile(0.75)).astype("int").astype("category")
        )

# add a category that randomly assigns a letter A, B, C, D to each row with probability mass conditional
# on the target category
np.random.seed(42)


def target_1_cat(prob):
    if prob < 0.05:
        return "A"
    elif prob < 0.15:
        return "B"
    elif prob < 0.3:
        return "C"
    else:
        return "D"


def target_0_cat(prob):
    if prob < 0.05:
        return "D"
    elif prob < 0.15:
        return "C"
    elif prob < 0.3:
        return "B"
    else:
        return "A"


df["rand_ind"] = np.random.random(df.shape[0])
df["random_cat"] = (
    df["target rand_ind".split()]
    .apply(
        lambda df: target_1_cat(df.rand_ind)
        if df.target == 1
        else target_0_cat(df.rand_ind),
        axis=1,
    )
    .astype("category")
)

df.head()
```
